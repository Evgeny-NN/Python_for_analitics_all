### Условие 3: 3 задача
* Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
* Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
# создание графика
plt.figure(figsize=(8, 4))
plt.barh(df['condition'], df['price'])

# форматирование оси x
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
plt.gca().xaxis.set_major_formatter(tick)

# поворот меток оси x
plt.xticks(rotation=90)

# добавление подписей
plt.title('Распределение Состояний домов по цене')
plt.xlabel('Стоимость')
plt.ylabel("Оценка состояния")

# установка меток оси y
plt.yticks(df['condition'])

plt.show()
```

Из диаграммы видим, что состояние домов, наилучшее в ценовой категории 3,5 млн.$; самые дорогие дома имеют среднее состояние, 3-4 балла по пяти-бальной шкале, высокая цена таких домов вероятнее всего вырисовывается из других параметров, влияющих на цену.

```python
df.head(2)
```

```python
import seaborn as sns
corr_matrix = df[['bedrooms', 'floors', 'price', 'waterfront', 'grade', 'view', 'yr_built', 'lat', 'condition']].corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Из диаграммы видим, что наибольшее прямое влияние на стоимость дома оказывают признаки: состояния конструкции (0,7), в меньшей степени повышают стоимость кол-во спален, этажность и наличие набережной (0,3), плюс хороший вид (0,4).

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Прямая линейная зависимость цены и жилых квадратных метров. При увеличении жилого метража увеличивается стоимость.

```python
sns.jointplot(x=df['price'], y=df['sqft_living15'], kind='reg')
plt.title('Прилежащая жилая площадь соседей');
```

Можно предположить, что у домов с самыми высокими ценами от 4 млн. и выше снижается плотность соседской застройки. -> Дома без соседей.

```python
plt.figure(figsize=(4, 4))
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Прилежащая жилая площадь соседей');
```

Дома с видами 1-2 балла имеют одинаковую среднюю стоимость, 4-х бальные виды резко повышают цену домов
